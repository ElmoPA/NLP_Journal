```python
import numpy as np
```

```python
class Node:
    def __init__(self, dim):

        # the vector stored at this node
        self.data = np.random.randn(dim)
        
        #weights of how this node interacts with other nodes
        self.wkey = np.random.randn(dim, dim)
        self.wquery = np.random.randn(dim, dim)
        self.wvalue = np.random.randn(dim, dim)

    def key(self):
        # what do I have to do?
        return self.wkey @ self.data
    
    def query(self):
        # what am I looking for?
        return self.wquery @ self.data
    
    def value(self):
        # what do I publicity reveal/broadcast to others
        return self.wvalue @ self.data
```

```python
class Graph:
    def __init__(self):
        self.nodes = [Node(5) for _ in range(10)]
        
        randi = lambda: np.random.randint(len(self.nodes))
        self.edges = [[randi(), randi()] for _ in range(10)]
    
    def run(self):

        updates =[]
        for i, n in enumerate(self.nodes):
            #What is this node looking for?
            q = n.query()

            inputs = [self.nodes[ifrom] for (ifrom,ito) in self.edges if ito == i]
            if len(inputs) == 0:
                continue

            #gather the keys
            keys = [m.key() for m in inputs]
            #calculate the compatibilities
            scores = [k.dot(q) for k in keys]
            #softmax so that they sum to one
            scores = np.exp(scores)
            scores = scores/np.sum(scores)
            #gather the appropriate values with weighted sum
            values = [m.value() for m in inputs]
            update = sum([s * v for s, v in zip(scores, values)])
            updates.append(update)
            
        for n,u in zip(self.nodes, updates):
            n.data = n.data + u
```

```python
def show_node(n):
    print(f'''
        Data: {n.data}
        Key: {n.key()}
        Query: {n.query()}
        Value: {n.value()}
    ''')
```
